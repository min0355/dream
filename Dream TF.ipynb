#### 남들보다 뛰어난 감각을 동원하라 
    - 첫번째 버전은 언제나 실패작 : 아이디어와 창의성의 진보는 백지 위에 처음 밑그림을 그리는 순간과 그렸던 밑그림을 지우고 그 위에 다시 그리는 순간 사이에 존재함. ("아무것도 그리지 않은 상태에서 완성되는 것, 진부해지지 않으려면 그리지 않아야 한다." 이 말을 이해 할 수 있겠는가?)  

#### 창의력은 창의적 해석력, 독창성은 독창적 해석력 
    - 완전히 새로운 것을 만들어 낼 기회는 극히 드물다. 
    - 남들보다 더 잘 해석할 수 있는 감각을 선택하는 일이 중요한 이유 
    - 창의적 사람들은 끊임없이 "보는 법" 을 배우고 훈련하는 것 뿐  

#### 실패란 완전히 실패하는 것, 대체 불가능한 사명 찾기 
    - 실패는 하나의 사유로 되는 것이 아니다. 존재하는 모든 이유때문에 실패하고 완전한 실패만이 계속되는 실패를 멈추고 백지에서 다시 시작하게 한다.
    - 그래서 실패는 오래가지 않는다.
    - 트렌드는 중요하지 않다. 미래의 삶에 가장 중요한 역할을 하는 것은 '사명감' 이다. 
    - 다른 사람들이 엄두도 내지 못한 문제를 찾아 해결하는 노력이 사명감
    - 지금 떠올리는 문제 중 아직 아무도 해결하지 못한 것은 무엇인가? 
    - 독점 (소규모 시장에서 큰 폭의 점유율로 시작?), 비밀 (다른 사람은 미처 알아차리지 못한 독특한 기회 발견?), 유통 (고객에게 전달은?)   

#### 10 개 아이디어 생각이 안나면 20 개를 생각해라

    - 많은 것을 떠올리고 많은 것을 버려라. 
    - 당신이 할 수 있는 가장 미친 짓 리스트를 만들어라. 사람들은 당신의 진지하고 뛰어난 생각보다 미친 아이디어를 더 좋아할 가능성이 높다. 

#### 아이디어 만들기 연습
