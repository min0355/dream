PPP(GDP) : Purchasing Power Parity
구매력평가 지수를 기반으로 계산한 GDP를 GDP(PPP)라 한다. 각국의 통화단위로 산출된 GDP를 단순히 달러로 환산해 비교하지 않고 각국의 물가수준을 함께 반영하는 것이다.  
실제로 세계의 몇몇 나라들의 사례를 봤을 때 GDP도 낮고 GDP(PPP)도 그에 비해 낮은 경우, 그 국가의 평균적인 국민 생활상은 거의 멸망적인 수준이다.  
예를 들어 아메리카 대륙 및 카리브 해 인근 섬 전체를 통틀어 거의 북한 수준으로 압도적인 가난함을 자랑하는 아이티의 경우 2017년 1인당 명목 GDP는 761$인데, 
1인당 GDP(PPP)의 경우 1,810$로 GDP(PPP)/명목 GDP량이 2.3 정도이다.   

이는 1인당 명목 GDP가 비슷한 나라들에 비하면 크게 낮은 수치로 남아시아의 최빈국 중 하나인 방글라데시의 1인당 명목 GDP는 1,532$이고   
1인당 GDP(PPP)는 4,502$로 GDP(PPP)/명목 GDP량이 거의 3에 근접한다. 실제로 아이티의 체감 물가는 상당히 비싸다.   
아이티 근로자들의 월급이 30$(한화로 약 ￦36,000) 수준에 불과한데, 자체적인 공산품 생산 설비가 워낙 열악해 비누 같은 생필품을 거의 대부분 이웃의 상대적으로 부국인 
도미니카 공화국이나 해외에서 수입해 온다. 수입품의 경우 국제무역의 특성상 가난한 나라라고 해서 특별히 가격이 떨어지거나 하지 않는다.[5] 따라서 아이티인들은,   
예를 들어 도미니카 공화국에서 수입한 생필품의 경우 도미니카 공화국 사람들과 같은 값을 주고 사용해야 한다.   
도미니카 공화국에서 하루 일하고 번 돈으로 샴푸 10통을 살 수 있다고 하면, 아이티인들은 10일을 일해야 겨우 같은 샴푸 1통을 살 까 말까한 수준이다.   
따라서 아이티인들은 지독한 임금체불과 폭력에도 불구하고 어떻게든 옆 나라인 도미니카 공화국에서 일을 하러 떠나고, 더러는 미국까지 흘러들어가기도 한다.  
하지만 방글라데시의 경우, 절대 부유하다고는 할 수 없겠지만, 적어도 인력과 공장들 위주로 굴러가는 경제 덕분에 공산품이 그렇게 비싸지 않은 데다 주변국인 인도나 미얀마와의  
소득 격차도 그다지 크지 않아 아이티만큼 살인적인 물가에 시달리지는 않는다. 방글라데시인 평균 월급이 5만 원 정도인 건 안 비밀  

한가지더 쉽게 예시를 들자면 PPP를 자동차에 비유하자면 1인당 GDP(명목)가 낮지만 1인당 GDP(PPP)가 높은 경우는 양산형 대중차에 해당되고 1인당 GDP(명목)가 높고 
1인당 GDP(PPP)가 낮은 경우는 프리미엄 브랜드라고 볼 수 있는데 둘의 공통점은 어떻게든 작동하고 굴러가지만 브랜드 가치의 차이로 인해 가격의 차이가 나는 정도라고 볼 수 있다.  
[6] 고로 1인당 GDP의 명목값이 높은 국가일수록 해당 국민의 실질적인 소비가 고급, 명품을 소비하는 성향이 있고 1인당 GDP의 PPP값만 높은 국가들의 경우 이냥저냥 생필품에   
만족하고 살지만 수입 공산품은 비싸서 품질 낮고 싸구려인 물품을 쓸 가능성이 높다.[7]

    - 실패는 하나의 사유로 되는 것이 아니다. 존재하는 모든 이유때문에 실패하고 완전한 실패만이 계속되는 실패를 멈추고 백지에서 다시 시작하게 한다.

    - 다른 사람들이 엄두도 내지 못한 문제를 찾아 해결하는 노력이 사명감
    - 지금 떠올리는 문제 중 아직 아무도 해결하지 못한 것은 무엇인가? 
    - 독점 (소규모 시장에서 큰 폭의 점유율로 시작?), 비밀 (다른 사람은 미처 알아차리지 못한 독특한 기회 발견?), 유통 (고객에게 전달은?)

## INSIGHT (MIT)

#### 본질을 보기 어려운 정보 과잉의 시대에 살고 있다.
#### 그래서, 성과는 결국 통찰에 달려 있다.   
#### 사람은 의외로 깊이 생각하지 않고, 뇌는 습관대로 생각하므로 아래 9 가지 통찰을 방해하는 요소를 인지하고 있자. 
    - 현상의 반대를 결론으로 삼는 습관 
        1) 피곤하다 -> 휴가 쓰자
        2) 제품이 잘 안팔린다 -> 어떻게든 팔아야 한다. 
    - 일반론에 만족하는 습관 
        1) 살이 쪘다 -> 식사량을 줄이자
    - 프레임워크에 의존하는 습관
        1) SWOT (강점, 약점, 기회, 위협 매트릭스) 분석 후 생각을 그만하는 것 
           => 작업 자체에 성취감을 느끼고 깊이 생각했다는 착각에 빠짐. 
    - 카테고리에 빠지는 습관
        1) 그 사람 논리적이야 -> 이공계 출신일껄
        2) 요즘 A 사 실적이 별로래 -> 제조업이니 어쩔 수 없지 
           (그 와중에도 잘 나가는 제조업이 있다.)
    - 키워드에서 생각 멈추는 습관 
        1) 차별화하자, 경쟁 우위 확보, 고객만족도 향상 등 키워드 하나로 정리하면 
           구체성이 사라짐.
    - 초기 가설을 고집하는 습관
        1) 가설은 본래 새로운 정보나 발견에 의해 진화하는데 초기 가설만 고집하면 
           진화가 원천 차단됨. 
        2) 한 지역에서 성공한 사례도 다른 지역에선 가설이 달라질 수 있다. 
        2) 토론이나 회의 시 "다만", "하지만", "그러나" 등의 말을 많이 쓰는 증상을 보임.
    - 생각의 입구에 머무는 습관
        1) 정보 수집 시 너무 몰입해 목적을 잊고 그것 자체에 집중하는 습관 
           (현상의 표층에서 머무를 수 밖에) 
        2) 분석을 위한 분석
    - 프로세스로만 돌리려는 습관
        1) 고객이 정말 원하는 것은 무엇인가? 
           -> 앙케이트 조사를 하면 고객이 원하는 것을 알 수 있습니다. 
           (프로세스에 의존한 답변)
        2) 1) 번의 옳은 대답은 "고객이 정말로 원하는 것은 대응의 내용이 아니라 
           신속한 대응이라 생각합니다. 그것은 고객 앙케이트를 통해 명확해 질 것입니다."
           와 같이 자신의 머리로 해결책을 제시하는 대답임. 
    - 주체성을 잃어버리는 습관
        1) 다른 사람의 머리에 의지하는 습관 
            * A 제품이 안 팔리는 이유는 고객이 이미 질렸기 때문이라 생각합니다. 
            * 내 생각은 그렇지 않네. 경쟁사 B 제품은 시장 점유율을 크게 확대하고 있지 않나? 
            * 그렇다면 타사의 C 제품도 고전을 면치 못하고 있겠군요.
